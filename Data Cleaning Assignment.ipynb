## 1. Some values in the the FlightNumber column are missing. These numbers are meant to increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in these missing numbers and make the column an integer column (instead of a float column).

```python
df.dtypes
```

```python
# Checking Null Values:
df['FlightNumber'].isnull()
```

```python
# Trying to find the indexes of Null values:
arr = np.where(df['FlightNumber'].isnull()==True)
arr
```

```python
type(arr)
```

```python
# Since arr is tuple, therefore accessing the first element of tuple i.e. accessing only array
for i in arr[0]:
    print(i)
```

```python
# df.loc[Row, column]
# df.loc[3,'FlightNumber']
```

```python
# Replacing NUll values with increment of 10 in previous value: using the indexes we found in arr!
for i in arr[0]:
    df.loc[i,'FlightNumber'] = df.loc[i-1,'FlightNumber']+10
   
```

```python
df
```

```python
# Changing Datatype to int from float:
df=df.astype({'FlightNumber':int})
```

```python
df
```

```python
# Second way to change datatype:
df['FlightNumber'] = df['FlightNumber'].apply(lambda a: int(a))
```

```python
df
```

## 2. The From_To column would be better as two separate columns! Split each string on the underscore delimiter _ to give a new temporary DataFrame with the correct values. Assign the correct column names to this temporary DataFrame.

```python
df
```

```python
temporarydf = df.copy()
```

```python
lst1= []
lst2 = []
for i in temporarydf['From_To']:
    temp = i.split('_')
    lst1.append(temp[0])
    lst2.append(temp[1])
```

```python
lst1
```

```python
temporarydf['From'] = pd.DataFrame(lst1)
temporarydf['To'] = pd.DataFrame(lst2)
temporarydf

# OR: temporarydf[['From','To']] = temporarydf.From_To.str.split('_', expand = True)
```
